For a diatomic molecule like HCl, **we shall determine the *vibrational frequency, $\nu_e$, the rotational constant, $B_e$, and the vibrational correction term, $\alpha_e$.*** We can then calculate the equilibrium distance $R_e$ (H–Cl bond length) and the bond force constant $k$, which characterizes the strength of the H–Cl chemical bond. 

## 2 Theory: Vibrations and rotation of a diatomic molecule

The description of Sec. 1 is based on ideas from the classical chemical structure theory, which assumes the existence of chemical bonds of certain rigidity, partial atomic charges, etc. Quantitative description of molecular systems, however, is given by the laws of quantum mechanics. The solution of the Schrödinger equation provides the rotational-vibrational energy levels $E_{rv}$ which depend on the vibrational quantum number v = 0, 1, 2, ... and the rotational quantum number $J = 0, 1, 2, ...$,

$$
E_{rv}(v, J) = hB_eJ(J + 1) + h\nu_e
\left(
v + \frac{1}{2}
\right)
\tag{1}
$$

See Eqs. 17.76 and Eq. 17.81 in the Book of Levine. Here,

$$
B_e = \frac{h}{
8\pi^2 \mu R_e^2} \quad
\text{and} \quad \nu_e = \frac{1}{2\pi}
\sqrt{\frac{k}{\mu}}
\tag{2}
$$

are the equilibrium rotational constant and harmonic frequency, respectively. Other parameters are the
bond force constant k and the reduced mass

$$
\mu = \frac{M_H M_{Cl}}
{M_H + M_{Cl}}
\tag{3}
$$

As follows from Eq. (1), rotational-vibrational energies of a diatomic molecule are quantized : only
certain discrete levels are allowed by quantum mechanics. These levels are not directly observable though;
in any sort of experiment we can only see the transitions between them, which results in absorption or
emission of energy leading, correspondingly, to either absorption or emission spectra

### 2.1 Selection rules
```{figure} images/selection_rules.png
---
width: 600px
---
Energy level diagram and allowed transitions in a rotational-vibrational spectrum of a diatomic molecule. The spacing between vibrational levels (∆v = 1) is large compared to those between rotational ones (∆J = ±1).
```

If all possible transitions between energy levels were allowed, a hypothetical spectrum would be quite cluttered. Fortunately, only certain transitions are visible in a spectrum. Allowed transitions are determined by the physical mechanism of interaction between radiation and matter; as was mentioned in Sec. 1, a diatomic molecule has to be polar to show an IR absorption rotational-vibrational spectrum. In terms of its quantum numbers, selection rules of the IR spectrum of a diatomic molecule are

$$
\Delta v = \pm 1, \Delta J = ±1. \tag{4}
$$

Allowed transitions are conveniently represented on the energy diagram, Fig. 3. At room temperature, the energy spacing between vibrational levels of HCl is large compared to the characteristic thermal energy, and the majority of molecules are at their ground vibration level, so that the absorption spectrum corresponds to a ∆v = +1 transition (from the ground vibrational level to the ﬁrst excited state).

```{figure} images/ftir_spectra_labels.png
---
width: 600px
---
Rotational-vibrational infrared spectrum of HCl. Note the absence of a Q branch, which would have been located in between P and R branches but is forbidden by selection rules.
```

By convention, a series of transitions for which ∆J = −1 is called a P branch, and a series of lines for which ∆J = +1 is called an R branch. An R branch transition originating in state J and terminating in state (J + 1) is written as R(J), and a P branch transition originating in state J and terminating in state (J − 1) is written as P (J). Q branch, in which ∆J = 0 can occur in some cases, but not in the rotational-vibrational spectra of diatomic molecules, so we will ignore them.


### 2.2 IR spectrum structure
A recorded IR spectrum for a HCl molecule is plotted in Fig. 4. It has the expected structure: there are two series of lines, P and R branches, with a gap between them. This gap, which is a Q branch, would correspond to a pure vibrational (∆v = 1, ∆J = 0) transition with a frequency $\nu_e$—if it were not forbidden. We observe also that distance between peaks is not quite constant as one would expect from Eq. (1)! For example, for the R branch

$$
R(J) = E_{rv}(1, J + 1) − E_{rv}(0, J) = h \nu_e + 2hB_e(J + 1), \tag{5}
$$

and the distance between neighboring peaks $R(J + 1) − R(J) = 2hB_e$ should be a constant. The reason for this variable spacing is the following: Typically, a molecular vibration takes on the order of $10^{-14}$ s. A molecular rotation is normally much slower, taking $10^{−9}$ s to $10^{−10}$ s. Hence, as a molecule rotates one revolution, it vibrates many times. To account for this, we have to replace $R_e$ in the deﬁnition of the rotational constant, Eq. (2), with a vibrationally averaged $R$. In the simplest approximation, this leads to the v-dependence of the rotational constant,

$$
B_v = B_e − \alpha_e
\left(
v + \frac{1}{2}
\right)
\tag{6}
$$

For example, $B_0$ is the rotational constant for $v = 0$ and $B_1$ is the rotational constant when $v=1$. The rotational-vibrational energy (1) has to be modiﬁed accordingly,

$$
E_{rv}(v, J) = hB_v J(J + 1) + h\nu_e
\left(
v + \frac{1}{2}
\right)
\tag{7}
$$

This equation provides the basis for analyzing the vibration-rotation spectrum of HCl.



## 3 Spectrum Analysis
Using Eq. (7), we can derive an expression for frequencies in the R and P branches

$$
\begin{align}
R(J) &= E_{rv}(1, J + 1) − E_{rv}(0, J) = h\nu_e + 2hB_e(J + 1) − hα_e(J + 1)(J + 3),
\tag{8} \\
P (J) &= E_{rv}(1, J − 1) − E_{rv}(0, J) = h\nu_e − 2hB_eJ − hα_eJ(J − 2).
\tag{9}
\end{align}
$$

Equations (8) and (9) can be simpliﬁed further to assist you in determining $\nu_e$, $B_e$, and $\alpha_e$ from your assigned spectral frequencies. Substituting $n = J + 1$ in Eq. (8), and $n = −J$ in Eq. (9), we combine them together in a single equation

$$
\nu_{R,P}(n) = \nu_e + 2 (B_e − α_e) n − α_e n^2. \tag{10}
$$

We can now plot the results for both branches of the spectrum on the same graph. Fitting data to a non-linear regression (of a particular functional form you might recognize) by Eq. (10) will yield us several quantities that allow us to calculate our desired physical characteristics of the molecule and allowed transitions.

Alternatively, we can take the diﬀerence between pairs, rather than the spectral frequencies themselves, to simplify your data plot,

$$
∆\nu_{R,P}(n) = ν_{R,P}(n + 1) − ν_{R,P}(n) = (2B_e − 3α_e) − 2α_en.
\tag{11}
$$

Fitting data by Eq. (11) with a linear regression will give us $−2\alpha_e$ as a slope and $(2B_e − 3\alpha_e)$ from the intercept.
